```
import vega_datasets
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn import linear_model
import sklearn.metrics as sm
```

```
cr = vega_datasets.data.cars()
sn.pairplot(cr,size=1.5)
```

```
#cek apa ada data kosong ?

set(cr['Displacement'].isna())
```

```
set(cr['Miles_per_Gallon'].isna())
```

```
set(cr['Horsepower'].isna())
```

```
cr['Miles_per_Gallon'] = cr['Miles_per_Gallon'].fillna(cr['Miles_per_Gallon'].mean()) #isi data kosong dengan mean
cr['Horsepower'] = cr['Horsepower'].fillna(cr['Horsepower'].mean())
```

```
set(cr['Miles_per_Gallon'].isna())
```

```
set(cr['Horsepower'].isna())
```

```
lin_reg = linear_model.LinearRegression()
lin_reg.fit(cr[['Displacement','Horsepower','Weight_in_lbs']],cr['Miles_per_Gallon'])
```

```
cr['Displacement'].describe()
```

```
cr['Horsepower'].describe()
```

```
cr['Horsepower'].describe()
```

```
#coba prediksin nilai miles per gallon dengan variable displacement, horsepower, dan weight in lbs
lin_reg.predict([[100,200,4000]])
```

```
lin_reg.coef_
```

```
lin_reg.intercept_ #int ->intercept
```
